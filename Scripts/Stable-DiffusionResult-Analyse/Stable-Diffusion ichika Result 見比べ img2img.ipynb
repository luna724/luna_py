```python
#@markdown Template Text Format Creator
print("If Not Changed.\nScript are importing From latest Markdown")
rsf = ""
tilling = ""
oas = ""
hfix = ""
x = input("1. Part Number: Part.")
part = str(x)
x = input("2. Paste Prompt: ")
prompt = str(x)
x = input("3. Paste Negative Prompt: ")
negative = str(x)
x = input("4. Batch Size: ")
batch_size = str(x)
x = input("5. Batch Count: ")
batch_count = str(x)
x = input("6. CFG Scale: ")
CFG = str(x)
x = input("7. Seed: ")
seed = str(x)
x = str(input("8. Restore Faces (0 or 1) (False or True)"))
if x == "0":
    rsf = "False"
elif x == "1":
    rsf = "True"
elif x == "False" or "True":
    rsf = str(x)
x = str(input("9. Tilling (0 or 1) (False or True)"))
if x == "0":
    tilling = "False"
elif x == "1":
    tilling = "True"
elif x == "False" or "True":
    tilling = str(x)
x = input("10. Sampling Method (Name): ")
sm = str(x)
x = input("11. Sampling Step: ")
ss = str(x)
x = str(input("12. You Write Other Advanced Settings?  (0 or 1): "))
if x == "0":
    overwrite = False
    oas = "Default"
elif x == "1":
    overwrite = True
    oas = str(input("""12. Enter Other Advanced Settings: \n"""))
extension = input("Input Source Extension (jpg, png, jpeg):")
x = input("13. Stable Diffusion Checkpoint: ")
if x == "":
    sdcp = ""
else:
    sdcp = f"Stable Diffusion Checkpoint: {x}\n        "
x = input("14. Resize Setting (to or by): ")
if x == "to":
    x = input("14. Enter Width: ")
    w = str(x)
    x = input("14. Enter Height: ")
    h = str(x)
    resolution = f"{w},{h}"
    res_pt = f"to {resolution}"
elif x == "by":
    x = input("Resize by: ")
    y = input("Resize from Width (W*H): ")
    rw = y
    y = input("Resize from Height (W*H): ")
    rh = y
    y = input("Resize To Width (W*H): ")
    rwb = y
    y = input("Resize To Height (W*H): ")
    rhb = y
    rbase = f"{rw},{rh}"
    rbaseto = f"{rwb},{rhb}"
    res_pts = f"by {str(x)}"
    res_pt = f"{res_pts} (from {rbase} to {rbaseto})"
x = input("15. Resize Mode (Name): ")
rsm = str(x)


# メイン生成
print(f"""
      \n\n\n
# Part {part}
    # Generate
        Prompt: {prompt}

        Negative Prompt: {negative} 
        
        Resize {res_pt}  |  Resize Mode: {rsm}
        Batch Count / Size = {batch_count},{batch_size}
        CFG Scale = {CFG}  |  Seed = {seed}
        Restore faces: {rsf}  |  Tilling: {tilling}
        Sampling method: {sm}   |   Sampling Steps: {ss}
        {sdcp}Other Advanced Settings: {oas}\n

    # 問題点 / Issue
        !Write Issue Here
        
    # 傑作だと思うもの / Masterpiece
        !Write Masterpiece Here
    
# Input Source
![Input Source](./ipynb/img2img/in/p{part}.{extension})
# Output Image
![Output Image](./ipynb/img2img/out/p{part}.png)
\n\n\n\n\n""")
```
